#### Age ranges
* bellow 18 age
* 19-64
* elder 60

```python
def make_age_group(row):
    if row['Age'] < 18:
        return 'Child'
    if row['Age'] < 64:
        return 'Adult'
    if row['Age'] != row['Age']:
        return 'Undefined'
    else:
        return 'Senior'
```

```python
make_age_group(tickets.loc[0, :])
```

```python
#axis = 1 -строки
tickets['age_group'] = tickets.apply(make_age_group, axis = 1)
```

```python
tickets.groupby['age_group'].agg({'Age': ['mean', 'min']})
```

## Sophisticated functions for data frames and working with links

### Link parsing

```python
ad_stats = pd.read_excel('ad_campaigns.xlsx')
ad_stats.head()
```

```python
ad_stats.rename(columns={'Название группы':'group','Фраза (с минус-словами)': 'phrase',

'Продуктивность': 'effect', 'ID объявления': 'ad_id', 'Заголовок': 'title',

'Текст': 'text', 'Ссылка': 'link'}, inplace = True)
```

```python
ad_stats.head()
```

```python
from urllib import parse
```

```python
url = ad_stats.loc[0, 'link']
```

```python
parsed = parse.urlsplit(url)
parsed
```

```python
print(parsed.netloc)
```

```python
query_string = parsed.query
query_string
```

```python
params_dict = parse.parse_qs(query_string)
params_dict
```

```python
params_dict['utm_campaign']
```

```python
params_dict['utm_campaign'][0]
```

```python
def campaign_name( row ):
    """Get campaign name from link inside row"""
    parsed = parse.urlsplit(row['link'])
    params_dict = parse.parse_qs(parsed.query)
    return params_dict['utm_campaign'][0]
```

```python
ad_stats['campaign'] = ad_stats.apply(campaign_name, axis = 1)
```

```python
ad_stats.head()
```

```python
ad_stats['hash'] = ad_stats.apply(lambda x: len(x['campaign']), axis = 1)
```

```python
ad_stats.head()
```

```python
ad_stats['hash'].mean()
```

# Unit 5

## Groupings

```python
ad_stats['campaign'].value_counts().head()
```

```python
ad_stats.groupby('campaign').count().head()
```

```python
ad_stats.groupby('campaign').count().reset_index().sort_values(by='group', ascending = False).head()
```

```python
ad_stats.groupby('campaign').sum().sort_values('effect', ascending =False).head()
```

```python
ad_stats.groupby('campaign').min().sort_values('effect').head()
```

### Getting multiple metrics

```python
ad_stats.groupby('campaign').agg(['min', 'max'])['effect'].head()
```

```python
ad_stats['phrase_length'] = ad_stats.apply(lambda x: len(x['phrase']), axis=1)
```

```python
ad_stats.head()
```

```python
ad_stats.groupby('campaign').agg({'effect': ['min', 'max'], 'phrase_length':'mean'}).head()
```

### Grouping by multiple columns

```python
ad_stats.groupby(['group', 'campaign']).count().sort_values('phrase',ascending = False).reset_index().head()
```

```python
ad_stats.groupby('campaign').agg({'effect':'sum'}).sort_values('effect', ascending=True).head()
```

```python
geo_data = {
      'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'],
    'Сибирь': ['новосибирск']
}
```

```python
def geo_class( row ):
    """This function classifies queries by region"""
    for region, city_list in geo_data.items():
        for city in city_list:
            if city in row['keyword']:
                return region
    return 'Undefined'
```

```python
#data from keywords.csv
data['region'] = data.apply(geo_class, axis=1)
```
