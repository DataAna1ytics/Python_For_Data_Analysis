# Unit 1

```python
import pandas as pd
from urllib import parse
```

```python
ad_stats = pd.read_excel('ad_campaigns.xlsx')
ad_stats.rename(columns={'Название группы':'group','Фраза (с минус-словами)': 'phrase',

'Продуктивность': 'effect', 'ID объявления': 'ad_id', 'Заголовок': 'title',

'Текст': 'text', 'Ссылка': 'link'}, inplace = True)

def campaign_name(row):
    """
    Get campaign name from link
    """
    
    parsed = parse.urlsplit(row['link'])
    params_dict = parse.parse_qs(parsed.query)
    
    return params_dict['utm_campaign'][0]

ad_stats['campaign'] = ad_stats.apply(campaign_name, axis = 1)
ad_stats.head()
```

```python
def effect_class(row):
    if row['effect'] <= 3:
        return 'Low'
    if row['effect'] <= 7:
        return 'Mean'
    return 'High'
```

```python
ad_stats['effect_class'] = ad_stats.apply(effect_class, axis = 1)
ad_stats
```

```python
ad_stats.pivot_table(index = 'campaign', columns = 'effect_class', values = 'effect',
                     aggfunc = 'count', fill_value = 0)
```

```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
len(ratings['userId'].unique())
```

```python
ratings['rating'].min()
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()[['userId','movieId']]
ratings_count.head()
```

```python
ratings_count.sort_values('movieId').head()
```

```python
film_fans_ratings_count = ratings_count[ ratings_count['movieId'] >= 100]
film_fans_ratings_count.sort_values('movieId').head()
```

```python
film_fans_user_ids = film_fans_ratings_count['userId'].tolist()
film_fans_user_ids
```

```python
fans_data = ratings[ ratings['userId'].isin(film_fans_user_ids)]
fans_data.head()
```

```python
min_max_df = fans_data.groupby('userId').agg(['min', 'max'])['timestamp']
```

```python
min_max_df.head()
```

```python
diff = (min_max_df['max']-min_max_df['min'])
lifetime = diff.mean()/(3600*24)
```

```python
data = pd.read_csv('ratings.csv')
data_pivot = data.pivot_table(index = ['movieId', 'rating'], values = 'timestamp', aggfunc = 'count').reset_index()
```

```python
data_pivot.head()
```

```python
data_pivot['sum'] = data_pivot.groupby('movieId')['timestamp'].transform(lambda x: sum(x))
#data_pivot['sum'] = data_pivot.groupby('movieId').timestamp.transform(lambda x: sum(x))
```

```python
data_pivot.head(15)
```

```python
data_pivot['share'] = data_pivot['timestamp']/ data_pivot['sum']
data_pivot.head(10)
```

```python
data_pivot[data_pivot['movieId'] == 1]['share'].sum()
```
